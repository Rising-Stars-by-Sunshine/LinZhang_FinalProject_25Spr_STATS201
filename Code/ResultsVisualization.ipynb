```python
import matplotlib.pyplot as plt
import numpy as np

# Define model names and their corresponding scores
models = [
    "Simulated_WEDDM", 
    "GSBlur_MPRNet", 
    "GSBlur_WEDDM_10Epochs", 
    "GSBlur_WEDDM_50Epochs"
]

psnr_scores = [31.0597, 8.8505, 20.3270, 21.1060]
ssim_scores = [0.9563, 0.2413, 0.7277, 0.7588]
lpips_scores = [0.0105, 0.6165, 0.2872, 0.2575]

# Define function to plot bar charts
def plot_scores(metric, scores, ylabel, title):
    plt.figure(figsize=(8, 6))
    colors = ['blue', 'red', 'green', 'purple']
    plt.bar(models, scores, color=colors, alpha=0.7)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.ylim(0, max(scores) * 1.2)  # Adjust y-axis range for better visibility
    plt.xticks(rotation=15)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Plot PSNR Comparison
plot_scores("PSNR", psnr_scores, "PSNR (Higher is Better)", "PSNR Comparison Across Models")

# Plot SSIM Comparison
plot_scores("SSIM", ssim_scores, "SSIM (Higher is Better)", "SSIM Comparison Across Models")

# Plot LPIPS Comparison
plot_scores("LPIPS", lpips_scores, "LPIPS (Lower is Better)", "LPIPS Comparison Across Models")
```
